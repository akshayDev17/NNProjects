# Explore the dataset
- Novel - Moby-Dick By Herman Melville.txt

```python
text_file_path = "/content/Novel - Moby-Dick By Herman Melville.txt"

chapter_counter, chapter_contents, chapters = 0, [], []

# read file line by line
with open(text_file_path, "r") as fp:
  for line in fp:
    if line != '\n':
      curr_line = line.replace('  ', ' ')
      curr_line = line.replace('\n', ' ')
      if curr_line.startswith("CHAPTER") or curr_line.startswith("Epilogue"):
        # start of new chapter

        # save contents of old chapter
        if len(chapter_contents) > 0:
          chapters.append(''.join(chapter_contents))
          chapter_contents = []

        print(f"Saved chapter {chapter_counter}.\n")

        # start temp-storing contents of new chapter
        chapter_counter += 1

        print(f"Started temp-storing contents of chapter {chapter_counter}.\n")
      else:
        chapter_contents.append(curr_line)

# Store epilogue
# save contents of old chapter
if len(chapter_contents) > 0:
  chapters.append(''.join(chapter_contents))
  chapter_contents = []

print(len(chapters))
```

# Model Definition

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, LSTM, Masking

model = Sequential()
model.add(Masking(mask_value=0, input_shape=(X.shape[-1],)))
model.add(Embedding(input_dim=y.shape[-1], output_dim=100)) # vocab size, embedding vector size, x_train each sequence size.
model.add(LSTM(150)) # dimensionality of h_t and c_t.
model.add(Dense(y.shape[-1], activation='softmax'))

print(model.summary())
```

If LSTM(150) is used, with embedding vector size = 100, then the reason for #parameters = 150600 is as follows:
1. There are 4 layers wherein $x_t$(of dimensionality 100) is converted to $c_t$(of dimensionality 150) - 4 matrices of dimensionality 100 x 150 $\Rightarrow$ 100 x 600 = 60,000.
2. Those 4 layers also convert $h_{t-1}$ to $c_t$ or $h_t$ - 4 matrices of dimensionality 150 x 150 $\Rightarrow$ 150 x 600 = 90,000.
3. Those 4 layers will each have 150 bias terms - 600 bias params.

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=20, validation_split=0.2)
```

# Stacked RNNs

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, SimpleRNN

stacked_rnn_model = Sequential()
stacked_rnn_model.add(Masking(mask_value=0, input_shape=(X.shape[-1],)))
stacked_rnn_model.add(Embedding(input_dim=y.shape[-1], output_dim=100)) # vocab size, embedding vector size, x_train each sequence size.
stacked_rnn_model.add(SimpleRNN(150, return_sequences=True))
stacked_rnn_model.add(SimpleRNN(100))
stacked_rnn_model.add(Dense(y.shape[-1], activation='softmax'))

print(stacked_rnn_model.summary())
```

```python
stacked_rnn_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
stacked_rnn_history = stacked_rnn_model.fit(X_train, y_train, epochs=20, validation_split=0.2)
```

## Double Stacking

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, LSTM

stacked_lstm_model = Sequential()
stacked_lstm_model.add(Masking(mask_value=0, input_shape=(X.shape[-1],)))
stacked_lstm_model.add(Embedding(input_dim=y.shape[-1], output_dim=150)) # vocab size, embedding vector size, x_train each sequence size.
stacked_lstm_model.add(LSTM(150, return_sequences=True))
stacked_lstm_model.add(LSTM(100))
stacked_lstm_model.add(Dense(y.shape[-1], activation='softmax'))

print(stacked_lstm_model.summary())
```

## 3-stacked LSTMs
