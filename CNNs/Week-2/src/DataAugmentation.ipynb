## Augmentation for a single image

```python
datagen = ImageDataGenerator(
    rotation_range = 40,
    shear_range = 0.2,
    zoom_range = 0.2,
    width_shift_range=0.2,
    height_shift_range=0.2
) # haven't kept horizontal_flip and vertical_flip since this changes the x-ray interpretation completely.
```

```python
import shutil

i = 0
output_dir = '/content/aug'

shutil.rmtree(output_dir)
os.mkdir(output_dir)

for output in datagen.flow(input_batch, batch_size=1, save_to_dir=output_dir):
  # happens to be an infinite iterator

  i += 1
  if i == 10:
    # generate 10 augmented images only
    break
```

## Augmentation when images are segregated into label-based directories

```python
def delete_train_datagen_directory(train_datagen_path):
  import shutil
  shutil.rmtree(train_datagen_path)
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1./255)

# read images in subdirs of train directory and generate augmented images
train_generator = train_datagen.flow_from_directory(
    './chest_xray/train',
    target_size=(150, 150),
    batch_size=16,
    class_mode='binary' # binary labels.
)

# no images need to be generated from validation data.
val_generator = test_datagen.flow_from_directory(
    './chest_xray/val',
    target_size=(150, 150),
    batch_size=1, # since there are only 16 images in validation dir
    class_mode='binary' # binary labels.
)
```

```python
from tensorflow.keras.layers import Conv2D, Flatten, Dense, AveragePooling2D, MaxPooling2D, Activation
from tensorflow.keras.models import Sequential

model = Sequential()

model.add(Conv2D(filters=32,kernel_size=(3,3),strides=(1, 1),input_shape=(150, 150, 3)))
model.add(Activation('relu'))

model.add(MaxPooling2D(pool_size=2,strides=2))

model.add(Conv2D(
    filters=32,
    kernel_size=(5,5),
    strides=(1, 1)
))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=2,strides=2))

model.add(Flatten())

model.add(Dense(64, activation="tanh"))
model.add(Dense(1, activation='sigmoid'))

print(model.summary())
```

## `fit_generator` using `train_generator`

```python
batch_size = 64

model.compile('rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(
    train_generator,
    steps_per_epoch=5216 // batch_size,
    epochs=20,
    validation_data=val_generator
)
```

## Observe results

```python
import matplotlib.pyplot as plt

plt.plot(range(len(history.history['loss'])), history.history['loss'], label='Training Loss')
plt.plot(range(len(history.history['loss'])), history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()
```

```python
plt.plot(range(len(history.history['accuracy'])), history.history['accuracy'], label='Training Accuracy')
plt.plot(range(len(history.history['accuracy'])), history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.show()
```

## Make predictions on Test Directory-data

```python
import numpy as np

test_dir = './chest_xray/test'
test_dir_images = []

for item in ['NORMAL', 'PNEUMONIA']:
  for x in os.listdir(os.path.join(test_dir, item)):
    test_dir_images.append(os.path.join(test_dir, item, x))

test_images_pil = [image.load_img(x, target_size=(150,150)) for x in test_dir_images]

test_images_arr = [image.img_to_array(x) for x in test_images_pil]

test_batch = np.array(test_images_arr)
```

```python
test_batch.shape
```

```python
# know to whom the train generator assigned what label out of 0 and 1
print(train_generator.class_indices)
```

```python
y_test = [0]*234
y_test += [1]*390
```

```python
predictions = (model.predict(test_batch) > 0.5).astype('int')

from sklearn.metrics import accuracy_score, confusion_matrix, f1_score

cm = confusion_matrix(y_test, predictions)

print(f"Accuracy = {accuracy_score(y_test, predictions)}, F1 score = {round(f1_score(y_test, predictions), 2)}")

# Display the confusion matrix
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()

classes = ['NORMAL', 'PNEUMONIA']  # Replace with your class names if applicable
tick_marks = np.arange(len(classes))

plt.xticks(tick_marks, classes)
plt.yticks(tick_marks, classes)

fmt = 'd'
thresh = cm.max() / 2.

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], fmt),
                 ha="center", va="center",
                 color="white" if cm[i, j] > thresh else "black")

plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.tight_layout()
plt.show()
```
