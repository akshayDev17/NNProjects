```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
```

```python
X, y = make_circles(n_samples=1000, noise=0.1, random_state=1)
```

```python
sb.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2)
```

```python
import tensorflow as tf

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

# Glorot Normal Initialization

```python
from keras.initializers import GlorotNormal
```

```python
model = Sequential()

model.add(Dense(units=32, activation='tanh', kernel_initializer=GlorotNormal(seed=42), input_shape=(X_train.shape[-1],)))
model.add(Dense(units=16, activation='relu', kernel_initializer=GlorotNormal(seed=42)))
model.add(Dense(units=1, activation='sigmoid', kernel_initializer=GlorotNormal(seed=42)))

print(model.summary())
```

```python
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train_tensor, batch_size=32, validation_split=0.2, epochs=100)
```

```python
print(history.history.keys())

plt.plot(range(len(history.history['accuracy'])), history.history['accuracy'], label='Accuracy')
plt.plot(range(len(history.history['val_accuracy'])), history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.show()
```
