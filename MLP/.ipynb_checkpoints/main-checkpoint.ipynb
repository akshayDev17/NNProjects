```python
import pandas as pd
import numpy as np
import sklearn.datasets as skds
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def step(x):
    return 1 if x>=0 else 0
```

```python
print(step(-1));print(step(0));print(step(12))
```

```python
class PerceptronLayer:
    def __init__(self, in_dim, weights=None):
        self.in_dim=in_dim
        self.activation = step
        if weights is None:
            # initialize weights as 1 X in_dim matrix
            self.weights=np.ones((in_dim,1))
        else:
            self.weights=weights
    def predict(self, X):
        n_samples, n_features = X.shape
        x_new = X.copy()
        if n_features < self.weights.shape[0]:
            # add column of ones to X
            x_new = np.hstack((x_new, np.ones((1,))))
        return self.activation(x_new.dot(self.weight))
```

```python
data = skds.load_breast_cancer(as_frame=True)
print(data.DESCR)
```

```python
data.frame.head(10)
```

```python
X, y = data.data, data.target
display(X.head(5));print(y)
```

```python
print(y.value_counts())
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=442)
print(x_train.shape, x_test.shape)
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(x_train, y_train)
```

```python
print(p.score(x_test, y_test))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

y_pred = p.predict(x_test)
print(f"Precision={round(precision_score(y_test, y_pred), 3)}, Recall={round(recall_score(y_test, y_pred), 3)}, F1={round(f1_score(y_test, y_pred), 3)}")
```
